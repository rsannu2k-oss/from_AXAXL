```python
import threading

# List of tables to create views for
tables_to_create_views = [
    "AddressHeader", 
    "AreaCodes",
    "CoverageData",
    "EstimatedPolicyInwards",
    #"InformaticaHistoricalRoe",
    "NameInvolvements",
    "Names",
    "NamesAdditionalDetail",
    "NamesLinkedToAddresses",
    "PolicyAuditLog",
    "PolicyDetails",
    "PolicySection",
    "PolicySectionDetail",
    "ProgramLink"
]

# Source and destination details
source_catalog = "xliidw_dev_lpl"
source_schema = "staging_genius"
dest_catalog = "dw_kc_dev"
dest_schema = "staging_genius_reports"

def create_view(table_name):
    view_name = f"{table_name}"
    sql_create_view = f"""
    CREATE OR REPLACE TABLE {dest_catalog}.{dest_schema}.{view_name} AS
    SELECT * FROM {source_catalog}.{source_schema}.{table_name}
    """
    spark.sql(sql_create_view)
    print(f"Created view {dest_catalog}.{dest_schema}.{view_name} referencing {source_catalog}.{source_schema}.{table_name}")

threads = []
for table_name in tables_to_create_views:
    t = threading.Thread(target=create_view, args=(table_name,))
    t.start()
    threads.append(t)

for t in threads:
    t.join()
```

```python
ihr_table = f"""CREATE OR REPLACE TABLE {dest_catalog}.{dest_schema}.InformaticaHistoricalRoe
as
SELECT 
HistoricalRoe.FKCompany,
HistoricalRoe.FKRateGroup,
HistoricalRoe.FKCcy,
HistoricalRoe.ExchangeRate,
CASE
WHEN CorpHistoricalROE.ExchangeRate = 0 THEN 0
ELSE HistoricalRoe.ExchangeRate / CorpHistoricalROE.ExchangeRate
END
AS CorporateRate
FROM {source_catalog}.{source_schema}.HistoricalRoe
LEFT JOIN
(SELECT
HistoricalRoe.FKCompany,
HistoricalRoe.FKRateGroup,
HistoricalRoe.ExchangeRate,
BaseCcyCompany.FKCcyCode
FROM {source_catalog}.{source_schema}.historicalroe
LEFT JOIN {source_catalog}.{source_schema}.BaseCcyCompany ON
HistoricalRoe.FKCompany = BaseCcyCompany.PKCompany
WHERE
HistoricalRoe.FKCcy = 'USD'
) AS CorpHistoricalROE ON
CorpHistoricalROE.FKCompany = HistoricalRoe.FKCompany AND
CorpHistoricalROE.FKRateGroup = HistoricalRoe.FKRateGroup"""

spark.sql(ihr_table)
```
